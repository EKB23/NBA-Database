```python
import csv

def read_csv_column(filename, column_index):
    values = []  # Empty list to store the values
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) > column_index:
                value = row[column_index]
                values.append(value)  # Append value to the list

    return values  # Return the list of values

# Usage example
filename = 'listofnames.csv'
column_index = 2  # Index of the column you want to read (0-based)
player_lname = read_csv_column(filename, column_index)
player_lname
```
