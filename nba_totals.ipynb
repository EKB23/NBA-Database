```python
driver = webdriver.Chrome(
    executable_path="/Users/ericr/chromedriver_win32/chromedriver"
    )
```

```python
years = list(range(1960, 2024))
```

```python
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}.html"

for year in years:
    url = team_stats_url.format(year)
    
    driver.get(url)
    driver.execute_script("window.scrollTo(1,10000)")
    time.sleep(18)
    
    with open("NBA/{}.html".format(year), "w+", encoding="utf-8") as f:
        f.write(driver.page_source)
```

```python
from bs4 import BeautifulSoup

dfs = []
for year in years:
    with open("NBA/{}.html".format(year), encoding="utf-8") as f:
        page = f.read()
    
    soup = BeautifulSoup(page, 'html.parser')
    player_table = soup.find_all(id="div_advanced-team")[0]
    player_df = pd.read_html(str(player_table))[0]
    player_df["Year"] = year
    dfs.append(player_df)
```

```python
nba_stats = pd.concat(dfs)
```

```python
nba_stats.to_csv("nba_advanced_stats.csv")
```
